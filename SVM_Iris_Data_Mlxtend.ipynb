```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVC
```

```python
iris = datasets.load_iris()
print(type(iris))
```

```python
X = iris.data[:, :2] # we only take the first two features. 
y = iris.target # We could avoid this ugly slicing by using a two-dim dataset
```

```python
print(X.shape)
print(y.shape)
```

```python
print(type(X))
print(type(y))
```

### Plotting SVM with different kernels

```python
kernels = ['linear', 'rbf', 'poly']
for i in kernels:
    model = SVC(kernel=i).fit(X, y)
    plot_decision_regions(X, y, clf=model, legend=2)

    # Adding axes annotations
    plt.xlabel('sepal length [cm]')
    plt.ylabel('petal length [cm]')
    plt.title(f'SVM on Iris with kernel {i}')
    plt.show()
```

### Plotting SVM with differnt gamma values for rbf kernel

gamma is a parameter for non linear hyperplanes. The higher the gamma value it tries to exactly fit the training data set

```python
gammas = [0.001,0.01,0.1,1,5,10,50,100]
for i in gammas:
    model1 = SVC(kernel='rbf', gamma=i).fit(X, y)
    plot_decision_regions(X, y, clf=model1, legend=1)

    # Adding axes annotations
    plt.xlabel('sepal length [cm]')
    plt.ylabel('petal length [cm]')
    plt.title(f'SVM with rbf kernel on Iris with gamma {i}')
    plt.show()
```

#### Conclusion

We can see that increasing gamma leads to overfitting as the classifier tries to perfectly fit the training data

### Plotting SVM with differnt C values for poly kernel

#### C parameter

C is the penalty parameter of the error term. It controls the trade off between smooth 
decision boundary and classifying the training points correctly.

```python
c = [0.1, 1, 10, 100, 1000]
for i in c:
    model2 = SVC(kernel='poly', C=i).fit(X, y)
    plot_decision_regions(X, y, clf=model2, legend=2)

    # Adding axes annotations
    plt.xlabel('sepal length [cm]')
    plt.ylabel('petal length [cm]')
    plt.title(f'SVM with poly kernel on Iris with C {i}')
    plt.show()
```

#### Conclusion 

Increasing C values may lead to overfitting the training data.

### Plotting SVM with differnt degree values for poly kernel

#### degree

Degree is a parameter used when kernel is set to 'poly'. 
It's basically the degree of the polynomial used to find the hyperplane to split the data.

```python
degrees = [0, 1, 2, 3, 4, 5, 6]
for i in degrees:
    model2 = SVC(kernel='poly', degree=i).fit(X, y)
    plot_decision_regions(X, y, clf=model2, legend=2)

    # Adding axes annotations
    plt.xlabel('sepal length [cm]')
    plt.ylabel('petal length [cm]')
    plt.title(f'SVM with poly kernel on Iris with degree {i}')
    plt.show()
```

#### Conclusion

Used degree=1 is the same as using a 'linear' kernel. Also, increasing this parameters 
leads to higher training times.
