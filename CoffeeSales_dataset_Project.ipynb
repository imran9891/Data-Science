```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Loading the dataset
df = pd.read_excel('Coffee.xlsx')
df.head()
```

```python
# 1. Read the dataset
df.info()
```

```python
# 2. Remove/handle the null values
df.isnull().sum()
```

```python
# 3. Depict different Market types and their count on a bar chart
print(df.Market.value_counts())
df.Market.value_counts().plot.bar()
plt.xlabel("Market Type")
plt.ylabel("Count");
```

```python
# 4. Depict different Product types and their count on a line chart
print(df["Product Type"].value_counts())
df["Product Type"].value_counts().plot()
plt.xlabel("Product Type")
plt.ylabel("Count");
```

```python
# 5. Depict different State and their count on a pie chart
print(df.State.value_counts())
df.State.value_counts().plot.pie(figsize=(6,6),autopct="%1.1f");
```

```python
# 6. Depict Market wise sum of Profit and depict the result on bar chart.
print(df.groupby("Market").sum()["Profit"])
df.groupby("Market").sum()["Profit"].plot.bar()
plt.xlabel("Market Type")
plt.ylabel("Total Profit per Market");
```

```python
# 7. Depict Product wise mean of Sales and depict the result on bar chart and line chart superimposed
print(df.groupby("Product").mean()["Sales"])
df.groupby("Product").mean()["Sales"].plot.bar(figsize=(7,7))
df.groupby("Product").mean()["Sales"].plot(figsize=(7,7))
plt.xlabel("Products")
plt.xticks(rotation=45)
plt.ylabel("Average Sales");
```

```python
# 8. Depict Profit vs Sales on a Scatter Chart where Product is 'Regular Espresso'
Sales = df[df.Product == "Regular Espresso"].Sales
Profit = df[df.Product == "Regular Espresso"].Profit
plt.scatter(Sales,Profit)
plt.xlabel("Sales")
plt.ylabel("Profit");
```

```python
# 9. Depict Budget Profit vs Budget Sales on a Scatter Chart where State is California and Product is 'Earl Grey'
x = df[(df.State == "California") & (df.Product=="Earl Grey")]["Budget Sales"]
y = df[(df.State == "California") & (df.Product=="Earl Grey")]["Budget Profit"]
plt.scatter(x,y)
plt.xlabel("Budget Sales")
plt.ylabel("Budget Profit");
```

```python
# 10. Depict Market wise mean of Budget Sales and mean of Sales in a stacked bar chart
mbs = df.groupby("Market").mean()["Budget Sales"]
s = df.groupby("Market").mean()["Sales"]
labels = ["Central","East","South","West"]
plt.bar(labels, mbs)
plt.bar(labels,s,bottom=mbs)
plt.xlabel("Market Segments")
plt.ylabel("MEAN");
```

```python
# 11. Plot the Frequency distribution of Total Expenses
df["Total Expenses"].plot.hist();
```

```python
# 12. Compare Frequency Distribution of Profit and Budget Profit on a histogram for the Oregon State
df[df.State == "Oregon"].Profit.plot.hist(figsize=(7,5))
df[df.State == "Oregon"]["Budget Profit"].plot.hist(figsize=(7,5))
plt.legend();
```

```python
# 13. Compare Frequency of COGS and Budget COGS on a histogram for the Eastern Market for the Product Cafe Mocha
df[(df.Market == "Eastern") & (df.Product == "Caffe Mocha")]
```

```python
# 14. Compare Sales vs Profit on a histogram for the Eastern or Central Market and for Decaf Type
df[((df.Market=="Eastern") | (df.Market=="Central")) & (df.Type=="Decaf")].Sales.plot.hist()
df[((df.Market=="Eastern") | (df.Market=="Central")) & (df.Type=="Decaf")].Profit.plot.hist()
plt.legend();
```

```python
# 15. Depict the quartile distribution of COGS where Sales is greater than 400
print(df[df.Sales>400].COGS.describe()[4:7])
df[df.Sales>400].COGS.describe()[4:7].plot.bar();
```

```python
# 16. Depict the quartile distribution of Budget Profit for Decaf Espresso 
print(df[df.Product=="Decaf Espresso"]["Budget Profit"].describe()[4:7])
df[df.Product=="Decaf Espresso"]["Budget Profit"].describe()[4:7].plot.bar();
```

```python
# 17. Find the Market, Product and Type where Sales is maximum 
df.loc[df[df.Sales==max(df.Sales)].index[:],['Market','Product','Type']]
```

```python
# 18. Find State and Market with Maximum Profit or Maximum Budget Profit
df.iloc[df[(df.Profit==max(df.Profit)) | (df["Budget Profit"]==max(df["Budget Profit"]))].index[:],[7,2]]
```

```python
# 19. Find Product, State and Market where Total Expenses are minimum
df.loc[df[df["Total Expenses"]==min(df['Total Expenses'])].index[:],["Product","State","Market"]]
```

```python
# 20. Statisically infer the State, Product and Market chosen to further expand/invest into the Coffee Business based on 
# Sales, Profit and Total Expenses
df.iloc[df[(df.Sales==max(df.Sales))|(df.Profit==max(df.Profit))|(df["Total Expenses"]==min(df["Total Expenses"]))].index[:],[7,4,2]]
```

```python
# 21. Statisically infer the State, Product and Market chosen to further withhold the Coffee Business based on Sales, Profit
# and Total Expenses.
df.iloc[df[(df.Sales==min(df.Sales))|(df.Profit==min(df.Profit))|(df["Total Expenses"]==max(df["Total Expenses"]))].index[:],[7,4,2]]
```

**Pandas Plots**

```python
# 22. Depict Market and Type wise sum of Budget Profit on a bar chart 
df.groupby(["Market","Type"])['Budget Profit'].sum().plot.bar()
plt.ylabel("Sum");
```

```python
# 23. Depict Product Type and State wise mean of Sales on a bar chart
df.groupby(["Product Type","State"])["Sales"].mean().plot.bar(figsize=(15,10))
plt.ylabel("Mean");
```

```python
# 24. Depict Product and Type wise sum of Budget COGS on a line Chart
df.groupby(['Product','Type'])["Budget COGS"].sum().plot()
plt.ylabel('Sum')
plt.xticks(rotation=45);
```

```python
# 25. Depict Product and Type wise sum of Budget Profit and Budget Sales on a line Chart
df.groupby(["Product","Type"])['Budget Profit','Budget Sales'].sum().plot();
plt.ylabel("Sum")
plt.xticks(rotation=45);
```

```python
# 26. Depict Market and Product Type based mean of Sales on a bar chart and on a stacked bar chart separately
msm = df.groupby('Market')['Sales'].mean()
ptsm = df.groupby('Product Type')['Sales'].mean()
labels=['Coffee','Espresso','Herbal Tea','Tea']
plt.bar(labels,msm)
plt.bar(labels,ptsm,bottom=msm);
```

```python
# 27. Find State and Market Size based mean of Sales and COGS where Market is not Central and Profit is over 350. 
# Depict the result on a line chart and stacked bar separately
cond = df[(df.Market!='Central')&(df.Profit>350)]
cond.groupby(["State","Market Size"])["Sales"].mean().plot()
cond.groupby(["State","Market Size"])["COGS"].mean().plot();
plt.xticks(rotation=45);
plt.figure()
a= cond.groupby(["State","Market Size"])["Sales"].mean()
b= cond.groupby(["State","Market Size"])["COGS"].mean()
labels=['(California,Major Market)','(Massachusetts,Major Market)','(Nevada,Small Market)','(New York,Major Market)']
plt.bar(labels, a)
plt.bar(labels, b, bottom=a)
plt.xticks(rotation=45);
```

```python
# 28. Find Product and Market Size based mean of Budget COGS and Inventory where Type is Regular and neither product or its
# Type is Espresso. Depict the result on a line chart and stacked bar separately
cond = df[(df.Type=='Regular')&(df['Product Type']!='Espresso')]
cond.groupby(['Product','Market Size'])['Budget COGS'].mean().plot()
cond.groupby(['Product','Market Size'])['Inventory'].mean().plot()
plt.xticks(rotation=35);
plt.figure()
a= cond.groupby(['Product','Market Size'])['Budget COGS'].mean()
b= cond.groupby(['Product','Market Size'])['Inventory'].mean()
labels=['(Amaretto,Major Market)','(Amaretto,Small Market)',
        '(Colombian,Major Market)','(Colombian,Small Market)',
        '(Darjeeling,Major Market)','(Darjeeling,Small Market)',
        '(Earl Grey,Major Market)','(Earl Grey,Small Market)',
        '(Green Tea,Major Market)','(Green Tea,Small Market)']
plt.bar(labels, a)
plt.bar(labels, b, bottom=a)
plt.xticks(rotation=45);
```

**Seaborn EDA**

```python
# 29. Depict Product based on Market Size on a count plot
sns.countplot(x= df.Product,hue=df['Market Size']);
plt.xticks(rotation=90);
```

```python
# 30. Depict State based on Market on a count plot
sns.countplot(x=df.State, hue=df.Market)
plt.xticks(rotation=90);
```

```python
# 31. Depict the Product Type vs Total Expenses on a Violin Plot based on Market
sns.violinplot(x=df['Product Type'],y=df['Total Expenses'],hue=df.Market);
```

```python
# 32. Depict Market vs Budget COGS on a boxplot
sns.boxplot(x=df.Market,y=df["Budget COGS"]);
```

```python
# 33. Depict Product vs Margin on a boxplot based on Market Size
sns.boxplot(y=df.Product,x=df.Margin,hue=df['Market Size']);
```

```python
# 34. Depict the Market vs Budget Sales on a Violin Plot based on Product Type where product is Colombian or Darjeeling
cond = df[(df.Product=='Colombian')|(df.Product=='Darjeeling')]
sns.violinplot(x=cond.Market,y=cond['Budget Sales'],hue=cond['Product Type']);
```

```python
# 35. Depict the Product vs Inventory on a Box Plot based on Product Type where State is California or Nevada
cond = df[(df.State=='California')|(df.State=='Nevada')]
sns.boxplot(x=cond.Product,y=cond.Inventory,hue=cond['Product Type'])
plt.xticks(rotation=90);
```

```python
# 36. Generate Pairplot for the Coffee dataset where Type is Regular including fields- Budget Sales, Budget Profit,
# Budget COGS, Sales, Profit and COGS
cond = df[df.Type=='Regular']
sns.pairplot(cond,vars=['Budget Sales','Budget Profit','Budget COGS','Sales','Profit','COGS']);
```

```python
# 37. Generate Pairplot for the Coffee dataset where Profit is over 250 based on Type including fields- Budget Sales,
# Budget Profit, Budget COGS, Sales, Profit and COGS
cond = df[df.Profit>250]
sns.pairplot(cond,vars=['Budget Sales','Budget Profit','Budget COGS','Sales','Profit','COGS'],hue='Type');
```

```python
# 38. Generate Pairplot for the Coffee dataset where Market is Western or Central based and Product Line is beans based on 
# Type including fields- COGS, Sales, Proift and Total Expenses
cond = df[((df.Market=='Central')|(df.Market=='Western'))&(df['Product Line']=='Beans')]
sns.pairplot(cond, vars=['COGS','Sales','Profit','Total Expenses'],hue='Type');
```

```python
# 39. Depict Market vs Total Expenses based on Type on a Swarm Plot where Sales is over 200 and Market Size is Small
cond = df[(df.Sales>200)&(df["Market Size"]=='Small Market')]
sns.swarmplot(x=cond['Market'],y=cond['Total Expenses'],hue=cond.Type);
```

```python
# 40. Depict correlation values on a heatmap
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(),cbar=False,annot=True,fmt='.2f');
```
