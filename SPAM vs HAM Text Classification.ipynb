```python
# Removing punctuations
for i in strings:
    r = re.sub('[^a-zA-Z]',' ',i)
    # Word Tokenize
    tokenized_text = word_tokenize(r)
    # Removing stopwords
    sw = stopwords.words('english')
    swr_text = [i for i in tokenized_text if i not in sw]
    # Applying LancasterStemmer
    lc = LancasterStemmer()
    stemmed_text = [lc.stem(i) for i in swr_text]
    print(stemmed_text)
```

```python
# Removing punctuations
for i in strings:
    r = re.sub('[^a-zA-Z]',' ',i)
    # Word Tokenize
    tokenized_text = word_tokenize(r)
    # Removing stopwords
    sw = stopwords.words('english')
    swr_text = [i for i in tokenized_text if i not in sw]
    # Applying LancasterStemmer
    lc = LancasterStemmer()
    stemmed_text = [lc.stem(i) for i in swr_text]
    # Lemmatizing 
    lm = WordNetLemmatizer()
    lematized_text = [lm.lemmatize(i) for i in stemmed_text]
    print(lematized_text)
```

```python
filtered_text = []
# Removing punctuations
for i in strings:
    r = re.sub('[^a-zA-Z]',' ',i)
    # Word Tokenize
    tokenized_text = word_tokenize(r)
    # Removing stopwords
    sw = stopwords.words('english')
    swr_text = [i for i in tokenized_text if i not in sw]
    # Applying LancasterStemmer
    lc = LancasterStemmer()
    stemmed_text = [lc.stem(i) for i in swr_text]
    # Converting list of words to list of sentences
    filtered_text.append(" ".join(stemmed_text))
print(filtered_text)    
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
cvec = CountVectorizer(max_features=3000)
x_filtered_text = cvec.fit_transform(filtered_text)
print(cvec.vocabulary_)
print(cvec.get_feature_names())
print(len(cvec.vocabulary_))
```

```python
X_cvec = x_filtered_text.toarray()
X_cvec.shape
```

```python
tfidfvect = TfidfVectorizer()
x_filtered_text_2 = tfidfvect.fit_transform(filtered_text)
print(tfidfvect.vocabulary_)
print(tfidfvect.get_feature_names()),
print(len(tfidfvect.vocabulary_))
```

```python
X_tfidf = x_filtered_text_2.toarray()
X_tfidf.shape
```

```python
y = df.Category
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_cvec,y,test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
X_train_, X_test_, y_train_, y_test_ = train_test_split(X_tfidf,y,test_size=0.2)
X_train_.shape, X_test_.shape, y_train_.shape, y_test_.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
modelcv = MultinomialNB()
modelcv.fit(X_train,y_train)
modelcv.score(X_test,y_test)
```

```python
cv_preds = modelcv.predict(X_test)
cv_preds[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test,cv_preds))
print(classification_report(y_test,cv_preds))
```
