### Pie Chart
1) It is used to represent precentage distribution<br>
2) It works on a single variable<br>

```python
marks = [80,90,75,63,85]
sub = ['Maths','Phy','Chem','CS','Eng']
print(sum(marks))
```

```python
pmat = marks[0]/sum(marks)
pphy = marks[1]/sum(marks)
pchem = marks[2]/sum(marks)
pcs = marks[3]/sum(marks)
peng = marks[4]/sum(marks)
print('PMat',pmat)
print('Pphy',pphy)
print('Pchem',pchem)
print('Pcs',pcs)
print('Peng',peng)
```

```python
plt.pie(marks,labels=sub,autopct="%2.0f%%")
plt.show()
```

```python
plt.pie(marks,labels=sub,autopct="%.3f%%",
        colors=['tan','orange','lightblue','gray','cyan'],
        explode = (0.1,0,0,0.2,0),shadow=True)
plt.show()
```

### Boxplot
It is used to represent quartile distribution

### Terminologies
1) Q1 - 25 percentile (Median of max and Q2)<br>
2) Q2 - 50 percentile (Median)<br>
3) Q3 - 75 percentile (Median of min and Q2)<br>
4) min - Q1 - 1.5*IQR<br>
5) max - Q3 + 1.5*IQR<br>
6) Outliers - >max or <min

```python
x = [12,3,4,5,6,7,14,25,8,6,9,10]
print(x)
print(len(x))
```

```python
plt.boxplot(x)
plt.grid()
plt.show()
```

```python
import numpy as np
```

```python
x = [12,3,4,5,6,7,14,25,8,6,9,10]
x.sort()
print(x)
print('Length of x',len(x))
```

```python
q0 = np.percentile(x,0)
q1 = np.percentile(x,25)
q2 = np.percentile(x,50)
q3 = np.percentile(x,75)
q4 = np.percentile(x,100)
print('Q0',q0) # Q0 is 0 percentile
print('Q1',q1) # Q1 is 25 percentile
print('Q2',q2) # Q2 is 50 percentile
print('Q3',q3) # Q3 is 75 percentile
print('Q4',q4) # Q4 is 100 percentile
```

```python
iqr = q3 - q1
print('IQR',iqr)
bmin = q1 - 1.5*iqr
bmax = q3 + 1.5*iqr
print('Min',bmin)
print('Max',bmax)
```

### Subplots
1) It is used to depict multiple charts in the same graph

```python
p = [1,2,3,4,5,6]
q = [4,5,3,2,5,7]
r = ['Delhi','Kanpur','Mumbai','Kolkata','Agra','Noida']
print(len(p),len(q),len(r))
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3)
ax1.plot(p,q,color='green',marker='o')  # line chart
ax2.bar(p,q,color='orange')  # bar chart
ax3.pie(p,labels=r,autopct='%.2f%%') # pie chart
plt.show()
```

```python
# plt.figure(figsize=(8,7))
fig, (ax1,ax2,ax3) = plt.subplots(3,1)
ax1.plot(p,q,color='green',marker='o')  # line chart
ax2.bar(p,q,color='orange')  # bar chart
ax3.pie(p,labels=r,autopct='%.2f%%') # pie chart
plt.show()
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(3,1,figsize=(8,7))
ax1.plot(p,q,color='green',marker='o')  # line chart
ax2.bar(p,q,color='orange')  # bar chart
ax3.pie(p,labels=r,autopct='%.2f%%') # pie chart
plt.show()
```

```python
graph, ((ax1,ax3),(ax2,ax4)) = plt.subplots(2,2,figsize=(8,8))
ax1.bar(p,q,color='green')
ax2.pie(p,labels=r,autopct="%.2f%%")
ax3.boxplot(p)
ax4.plot(p,q,color='maroon',marker='*')
plt.show()
```

```python
graph, (ax1,ax2) = plt.subplots(1,2,figsize=(8,5))
ax1.bar(p,q,color='green')
ax2.plot(p,q,marker="*")

ax1.set_title('Bar Chart')
ax2.set_title('Line Chart')

ax1.set_xlabel('p')
ax1.set_ylabel('q')

ax2.set_xlabel('p')
ax2.set_ylabel('q')

plt.show()
```
